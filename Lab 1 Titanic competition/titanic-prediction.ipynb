# Importing the data and pytorch


```python
df = pd.read_csv('../input/titanic/train.csv')
df
```

```python
import torch
```

```python
gender_map = {'male':1,'female':0}

df['Gender'] = df['Sex'].map(gender_map)
df
```

```python
del[df['Sex']]
```

```python
df['Cabin'].value_counts()
```

```python
mode = df['Embarked'].mode()
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df['Embarked'].unique()
```

```python
df['S'] = np.where(df['Embarked'] == 'S', 1, 0)
df['Q'] = np.where(df['Embarked'] == 'Q', 1, 0)
df['C'] = np.where(df['Embarked'] == 'C', 1, 0)
```

```python
mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean)
df['Age'].unique()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

# spliting data for descision tree

```python
df.columns
```

```python
target =  'Survived'
cols = ['Pclass','Age','SibSp', 'Parch','Fare', 'Gender','S', 'Q', 'C']

y = df[target] # Target variable

X = df[cols] # Features

X.describe()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42,stratify=y)
```

# Creating descision tree classifer

```python
dt = tree.DecisionTreeClassifier()
dt = dt.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

# Submitting the test

```python
test = pd.read_csv('../input/titanic/test.csv')

# Gender mapping
test['Gender'] = test['Sex'].map(gender_map)

# Fill missing Embarked
test['Embarked'] = test['Embarked'].fillna('S')

# One-hot encoding for Embarked
test['S'] = (test['Embarked'] == 'S').astype(int)
test['Q'] = (test['Embarked'] == 'Q').astype(int)
test['C'] = (test['Embarked'] == 'C').astype(int)

# Fill missing Age
test['Age'] = test['Age'].fillna(mean)

# Predict
test['Survived'] = dt.predict(test[cols])
```

```python
submission = test[['PassengerId', 'Survived']]
submission.to_csv('submission.csv', index=False)
```
